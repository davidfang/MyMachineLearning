```python
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
sns.set_style("whitegrid")


# 专治画图不显示
%matplotlib inline

# plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  # 中文字
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False
plt.axhline(0.75,color='green',lw=2,alpha=0.7)
plt.axvline(5,color='red',lw=2,alpha=0.7)

x = np.linspace(0,20,100)
y = np.sin(x)
plt.plot(x,y)
plt.fill_between(x,y,0)
plt.fill_between(x,0.75,1,where=y>0.75,color='red',alpha=0.5)
# plt.fill_betweenx(y,x<2)
plt.show()

```

```python
points = np.arange(-5, 5, 0.01)
xs, ys = np.meshgrid(points, points)
z = np.sqrt(xs ** 2 + ys ** 2)

plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")
plt.imshow(z, cmap=plt.cm.gray)
plt.colorbar()
```

```python
t = np.arange(100) # 时间
x = np.linspace(0,5*np.pi,100)
# x = np.arange(0,1,100)

y = np.sin(x )
s =   t * y # 距离
d =   np.cos(x)

# 分辨率参数-dpi，画布大小参数-figsize
# plt.figure(dpi=300,figsize=(24,8))
plt.figure(figsize=(24,8))


# plt.plot(x,np.sin(x),label="sin(x)")
# plt.plot(x,x,label="x")
plt.plot(x,y,label="y")
# plt.plot(x,s,label="s")
# plt.plot(x,s/t,label="s/t")
plt.plot(x,d,label="d")
plt.legend(loc='best')#图列位置，可选best，center等

plt.show()
```

```python
x = np.arange(100)
y1 = np.random.binomial(1,0.8,100)#产生二项分布
y2 = np.random.randn(100)#正态分布
y3 = np.random.normal(1,0.8,100)#高斯分布
# 分辨率参数-dpi，画布大小参数-figsize
# plt.figure(dpi=300,figsize=(24,8))
plt.figure(figsize=(24,8))
plt.plot(x,y1,label="y1 二项分布")
plt.plot(x,y2,label="y2 正态分布")
plt.plot(x,y3,label="y3 高斯分布")
plt.legend(loc='best')#图列位置，可选best，center等
plt.show()
```

```python
plt.figure(figsize=(24,8))
# plt.scatter(x,y1,label="y1 二项分布")
plt.scatter(x,y2,label="y2 正态分布")
plt.scatter(x,y3,label="y3 高斯分布")
plt.legend(loc='best')#图列位置，可选best，center等
plt.show()
```
